#  Boxplots (RCNN model) (Figure 10 )

```python
# sns.axes_style()
def figsize(scale):
    fig_width_pt = 426.79134                        # Get this from LaTeX using \the\textwidth
    inches_per_pt = 1.0/72.27                       # Convert pt to inch
    golden_mean = (np.sqrt(5.0)-1.0)/2.0            # Aesthetic ratio (you could change this)
    fig_width = fig_width_pt*inches_per_pt*scale    # width in inches
    fig_height = fig_width*golden_mean              # height in inches
    fig_size = [fig_width,fig_height]
    return fig_size

data=pd.read_csv("/home/zied/Desktop/lstm.csv",index_col=0)
# print (data.head, data.keys())

sns.set(rc={"figure.figsize": (12,6), "xtick.labelsize": 14,'legend.fontsize': 48.0, 'font.size': 48})

ax= sns.boxplot(data=data.T, width=0.7) 
ax.set_xlabel("Subject",fontsize=12)
ax.set_ylabel("Accuracy %",fontsize=12)

plt.xticks(plt.xticks()[0], rotation  =45)

#save it as eps 
ax.figure.savefig('/home/zied/Pictures/foof.eps', format='eps', dpi=1200)
```
